## attempts to categorize patterns to know what's popular



- higher case letter => "T"
- lower case letter => "t"
- digit => 1
- if new sentences starts with quotes => next caracter type
- other characters if so which one (",.!?:;()[])

````python
def replacer(car):
    import re
    if re.match("[A-Z]", car):
        return "T"
    if re.match("[a-z]", car):
        return "t"
    if re.match("[0-9]", car):
        return "1"
    return car
````

````python
def surrounding_replacer(surrounding):
    before, after = surrounding
    before = "".join(replacer(c) for c in before)
    after = "".join(replacer(c) for c in after)
    return (before, after)
````

````python
def before_iter_replaced(corpus="brown"):
    for surroundings in raw_surroundings(read_segmented_corpus_iter(corpus)):
        yield (surrounding_replacer(surroundings))[0]
        
def after_iter_replaced(corpus="brown"):
    for surroundings in raw_surroundings(read_segmented_corpus_iter(corpus)):
        yield (surrounding_replacer(surroundings))[1]

def full_pattern(corpus="brown"):
    for surroundings in raw_surroundings(read_segmented_corpus_iter(corpus)):
        yield (surrounding_replacer(surroundings))
    
````

````python
import matplotlib.pyplot as plt


def sort_dic_by_value_and_head(dico, count_min=75):
    sorted_tuples = sorted(dico.items(), key=lambda item: -item[1])
    return {k: v for index, (k, v) in 
            enumerate(sorted_tuples) if v > count_min}

def plot_tops(dico):
    dico = sort_dic_by_value_and_head(dico)
    plt.bar(dico.keys(), dico.values(), width=1, color='g')
````

````python
from collections import Counter
c1 = Counter(before_iter_replaced())
c2 = Counter(after_iter_replaced())
plt.figure(0)
plot_tops(c1)
plt.figure(1)
plot_tops(c2)
````

````python
from collections import defaultdict
c3_new = defaultdict(int)
for k, v in c3.items():
    noko = k[0] not in x_pattern_of_interest
    nok1 = k[1] not in y_pattern_of_interest
    if noko and nok1:
        c3_new[("other", "other")] += 1
        continue
    if noko:
        c3_new[("other", k[1])] += 1
    if nok1:
        c3_new[(k[0], "other")] += 1
````
