## attempts to categorize patterns to know what's popular



- higher case letter => "T"
- lower case letter => "t"
- digit => 1
- if new sentences starts with quotes => next caracter type
- other characters if so which one (",.!?:;()[])

````python
def replacer(car):
    import re
    if re.match("[A-Z]", car):
        return "T"
    if re.match("[a-z]", car):
        return "t"
    if re.match("[0-9]", car):
        return "1"
    return car
````

````python
def surrounding_replacer(surrounding):
    before, after = surrounding
    before = "".join(replacer(c) for c in before)
    after = "".join(replacer(c) for c in after)
    return (before, after)
````

````python
def before_iter_replaced():
    for surroundings in raw_surroundings(read_segmented_corpus_iter("brown")):
        yield (surrounding_replacer(surroundings))[0]
        
def after_iter_replaced():
    for surroundings in raw_surroundings(read_segmented_corpus_iter("brown")):
        yield (surrounding_replacer(surroundings))[1]
````

````python
from collections import Counter
c1 = Counter(before_iter_replaced())
c2 = Counter(after_iter_replaced())
````
